```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("Survey_Data_Collated_ALL_SITES.csv")
```

## Acres per Deer:

```python
sns.set(rc={"figure.figsize":(20, 10)}, font_scale = 2) 
p = sns.pointplot(x="year", y="dens", hue="site", data=data, scale=2)
plt.axhline(15)
plt.xlabel("Deer Density", size=30)
plt.ylabel("Year", size=30)
plt.legend(title = "Training Site", title_fontsize="30")
```

## Buck to Doe Ratios:

```python
sns.set(rc={"figure.figsize":(20, 10)}, font_scale = 2) 
p = sns.pointplot(x="year", y="bdrat", hue="site", data=data, scale=2)
plt.axhline(1)
plt.xlabel("Year", size=30)
plt.ylabel("Buck to Doe Ratios", size=30)
plt.legend(title = "Training Site", title_fontsize="30")
```

## Fawn to Doe Ratios:

```python
sns.set(rc={"figure.figsize":(20, 10)}, font_scale = 2) 
p = sns.pointplot(x="year", y="fdrat", hue="site", data=data, scale=2)
plt.axhline(1)
plt.xlabel("Year", size=30)
plt.ylabel("Fawn to Doe Ratios", size=30)
plt.legend(title = "Training Site", title_fontsize="30")
```
